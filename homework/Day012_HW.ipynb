### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('em')[-1].text,d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
news = {}
for d in soup.find(class_="part_list_2").find_all('h3'):
    clas = d.find_all('em')[-1].text
    title = d.find(class_="date").text + d.find_all('a')[-1].text
    if clas in news:
        news[clas] += title
    else:
        news[clas] = title
```

### ③ 哪一個類別的文章最多

```python
clasnum = {}
for d in soup.find(class_="part_list_2").find_all('h3'):
    clas = d.find_all('em')[-1].text

    if clas in clasnum:
        clasnum[clas] += 1
    else:
        clasnum[clas] = 1
        
max(clasnum)
```
