## 1. 取出今天所有的新聞

```python
from datetime import datetime, timedelta
import datetime
today = datetime.datetime.now()
offset = datetime.timedelta(days=-1)
re_date = (today + offset).strftime("%Y/%m/%d")
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")
    
import time
while True:
    browser.execute_script("window.scrollTo(0, 100000);")
    time.sleep(2)
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    if re_date in soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_="date").text:
        print('==== STOP ====')
        break

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    if re_date in d.find(class_="date").text :
        break
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 2. 取出現在時間兩小時內的新聞

```python
news = {}

import datetime
today = datetime.datetime.now()
offset = datetime.timedelta(hours=-2)
re_date = (today + offset).strftime("%Y/%m/%d %H:%M")

from datetime import datetime
for d in soup.find(class_="part_list_2").find_all('h3'):
    date = d.find(class_="date").text
    title = d.find_all('a')[-1].text
    time_d = datetime.strptime(date, "%Y/%m/%d %H:%M")
    time = time_d.strftime("%Y/%m/%d %H:%M")
    if  time > re_date:        
        news[date] = title
        print(date, title)
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")


from datetime import datetime, timedelta
yesterday_date = datetime.now() - timedelta(3)
yes_month = yesterday_date.strftime("%m")
yes_day = yesterday_date.strftime("%d")
three_days_15_clock = datetime.strptime(yesterday_date.strftime("%Y/%m/%d") + ' 15:00', '%Y/%m/%d %H:%M')
three_days_17_clock = datetime.strptime(yesterday_date.strftime("%Y/%m/%d") + ' 17:00', '%Y/%m/%d %H:%M')

from datetime import datetime, timedelta
import datetime
re_date = (three_days_15_clock).strftime("%Y/%m/%d")


month = int(yes_month)
day = int(yes_day)



from selenium.webdriver.support.ui import Select
selM = Select(browser.find_element_by_id("selM"))
selM.select_by_value(str(month))
selD = Select(browser.find_element_by_id("selD"))
selD.select_by_value(str(day))
browser.find_element_by_id('button').click()

import time
from datetime import datetime, timedelta
while True:
    browser.execute_script("window.scrollTo(0, 100000);")
    time.sleep(1)
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    new_date = soup.find_all(class_="part_list_2")[0].find_all('h3')[-1].find(class_="date").text
    if datetime.strptime(new_date, '%Y/%m/%d %H:%M') < three_days_15_clock:
        break

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

news={}
from datetime import datetime
for d in soup.find(class_="part_list_2").find_all('h3'):
    date = d.find(class_="date").text
    title = d.find_all('a')[-1].text
    time_d = datetime.strptime(date, "%Y/%m/%d %H:%M")
    
    time = time_d.strftime("%Y/%m/%d %H:%M")
    if  time_d > three_days_15_clock and time_d < three_days_17_clock:        
        news[date] = title
        print(date, title)
```
